```python
import pandas as pd
import numpy as np
# Load the CSV file
file_path = "C:\\Users\\harik\\Downloads\\food_coded.csv"
data = pd.read_csv(file_path)
```

```python
print(data.info())
print(data.describe())
```

```python
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
data['Gender'] = data['Gender'].astype('category')
print(data.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
print(data.describe())

# Distribution of GPA
plt.figure(figsize=(10, 6))
sns.histplot(data['GPA'], kde=True)
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()

# Gender distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Breakfast vs. GPA
plt.figure(figsize=(10, 6))
sns.boxplot(x='breakfast', y='GPA', data=data)
plt.title('GPA vs. Breakfast Consumption')
plt.xlabel('Breakfast Consumption')
plt.ylabel('GPA')
plt.show()

# Calories per day vs. GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories_day', y='GPA', data=data)
plt.title('Calories per Day vs. GPA')
plt.xlabel('Calories per Day')
plt.ylabel('GPA')
plt.show()
```
